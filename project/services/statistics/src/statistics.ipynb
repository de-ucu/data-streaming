```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
f, ax = plt.subplots(figsize=(18, 8))
languages = pd.read_json('/statistics/languages.txt', lines=True)['lang']
languages = languages.value_counts().reset_index()[:20]
languages.columns = ['lang', 'count']


g = sns.barplot(x='count', y='lang', data=languages, palette="pastel", edgecolor=".6")
g.set_ylabel("Languages", fontsize=20)
g.set_xlabel("Count", fontsize=20)
g.tick_params(labelsize=15)
```

```python
f, ax = plt.subplots(figsize=(18, 8))
entities = pd.read_csv('/statistics/entities.txt', sep='\t', names=["entity"])
entities = entities.value_counts().reset_index(name='counts')[:20]

g = sns.barplot(x='counts', y='entity', data=entities, palette="pastel", edgecolor=".6")
g.set_ylabel("entity", fontsize=20)
g.set_xlabel("counts", fontsize=20)
g.tick_params(labelsize=15)
```

```python
LABELS = {0: 'negative', 1: 'neutral', 2: 'positive'}

sentiments = pd.read_csv('/statistics/sentiments.txt', sep='\t', names=["sentiment"])
sentiments = sentiments.value_counts().reset_index(name='counts')
sentiments = sentiments.replace({ 'sentiment': LABELS }).set_index('sentiment')
sentiments.plot.pie(y='counts', x='sentiment', figsize=(10, 10));
```
